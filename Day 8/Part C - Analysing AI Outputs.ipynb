Write a Python program that prints a diamond pattern of asterisks.
The user inputs the number of rows for the upper half.
Include proper spacing and use nested loops only (no string multiplication tricks).

```python
n = int(input("Enter number of rows: "))

# Upper half
for i in range(1, n + 1):

    for j in range(n - i):
        print(" ", end="")

    for j in range(2 * i - 1):
        print("*", end="")

    print()

# Lower half
for i in range(n - 1, 0, -1):

    for j in range(n - i):
        print(" ", end="")

    for j in range(2 * i - 1):
        print("*", end="")

    print()
```

### Readability

The code is easy to understand because:

Separate loops are used for upper half and lower half

Variable names (i, j) are simple and consistent

Logical structure follows pattern-building conventions

However, readability could be slightly improved by adding comments or input validation.

Nested Loops vs String Tricks

The solution correctly uses nested loops to print characters one by one:

for j in range(...):
    print("*", end="")

This follows the requirement of not using string multiplication like:

print("*" * (2*i-1))

✔ Requirement satisfied.

Time Complexity

Let n be the number of rows in the upper half.

Total rows printed:

2n - 1

Each row prints up to O(n) characters.

Therefore overall complexity:

O(n²)

This is expected for pattern-printing problems.

4️⃣ Improved Version

Below is an improved implementation with input validation and clearer structure.

```python
n = int(input("Enter number of rows: "))

if n <= 0:
    print("Invalid input. Please enter a positive number.")

else:

    # Upper half
    for i in range(1, n + 1):

        for j in range(n - i):
            print(" ", end="")

        for j in range(2 * i - 1):
            print("*", end="")

        print()

    # Lower half
    for i in range(n - 1, 0, -1):

        for j in range(n - i):
            print(" ", end="")

        for j in range(2 * i - 1):
            print("*", end="")

        print()
```
