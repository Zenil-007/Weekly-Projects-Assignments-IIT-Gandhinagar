```python
# password_analyzer_generator.py

import string
import random


def analyze_password(password):
    """
    Analyze password strength based on length, character types,
    and repetition rules.
    """

    score = 0
    missing = []

    # ---- Length Check ----
    length = len(password)

    if length >= 16:
        score += 3
    elif length >= 12:
        score += 2
    elif length >= 8:
        score += 1
    else:
        missing.append("too short")

    # ---- Character Type Checks ----
    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False

    special_chars = "!@#$%^&*"

    for char in password:
        if char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
        elif char.isdigit():
            has_digit = True
        elif char in special_chars:
            has_special = True

    if has_upper:
        score += 1
    else:
        missing.append("uppercase")

    if has_lower:
        score += 1
    else:
        missing.append("lowercase")

    if has_digit:
        score += 1
    else:
        missing.append("digit")

    if has_special:
        score += 1
    else:
        missing.append("special character")

    # ---- Repeated Character Check ----
    repeated = False

    for i in range(len(password) - 2):
        if password[i] == password[i+1] == password[i+2]:
            repeated = True

    if not repeated:
        score += 1
    else:
        missing.append("no more than 2 repeated characters")

    return score, missing


def get_strength_rating(score):
    """Return textual rating based on score."""

    if score <= 2:
        return "Weak"
    elif score <= 4:
        return "Medium"
    elif score <= 6:
        return "Strong"
    else:
        return "Very Strong"


def generate_password(length):
    """
    Generate a random password using letters,
    digits, and punctuation.
    """

    characters = string.ascii_letters + string.digits + string.punctuation
    password = ""

    for _ in range(length):
        password += random.choice(characters)

    return password


def password_analyzer_loop():
    """
    Ask user for password until strength >= 5.
    """

    while True:

        password = input("Enter password: ")

        score, missing = analyze_password(password)

        rating = get_strength_rating(score)

        print(f"\nStrength: {score}/7 ({rating})")

        if missing:
            print("Missing:", ", ".join(missing))

        if score >= 5:
            print("Password accepted!\n")
            break
        else:
            print("Try again...\n")


def password_generator_section():
    """Generate password and analyze its strength."""

    length = int(input("Enter desired password length: "))

    generated = generate_password(length)

    print("\nGenerated Password:", generated)

    score, _ = analyze_password(generated)

    print("Strength Rating:", get_strength_rating(score))


# ---- Main Execution ----

password_analyzer_loop()
password_generator_section()
```
