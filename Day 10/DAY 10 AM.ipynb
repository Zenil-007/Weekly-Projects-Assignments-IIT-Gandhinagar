### 🧠 Part A — E-Commerce Product Catalog System

```python
# PART A — E-COMMERCE PRODUCT CATALOG SYSTEM

from collections import defaultdict

# Product catalog

catalog = {
    'SKU001': {'name':'Laptop','price':65000,'category':'electronics','stock':15,'rating':4.5,'tags':['computer','work']},
    'SKU002': {'name':'Phone','price':40000,'category':'electronics','stock':20,'rating':4.4,'tags':['mobile','android']},
    'SKU003': {'name':'Headphones','price':3000,'category':'electronics','stock':0,'rating':4.2,'tags':['music','audio']},
    'SKU004': {'name':'Tablet','price':25000,'category':'electronics','stock':10,'rating':4.1,'tags':['tablet','work']},

    'SKU005': {'name':'T-Shirt','price':900,'category':'clothing','stock':50,'rating':3.8,'tags':['casual','summer']},
    'SKU006': {'name':'Jeans','price':2000,'category':'clothing','stock':25,'rating':4.0,'tags':['denim','casual']},
    'SKU007': {'name':'Jacket','price':4500,'category':'clothing','stock':5,'rating':4.3,'tags':['winter','fashion']},
    'SKU008': {'name':'Sneakers','price':3500,'category':'clothing','stock':0,'rating':4.1,'tags':['shoes','sport']},

    'SKU009': {'name':'Clean Code','price':700,'category':'books','stock':30,'rating':4.7,'tags':['programming','software']},
    'SKU010': {'name':'Python Crash Course','price':850,'category':'books','stock':40,'rating':4.6,'tags':['python','coding']},
    'SKU011': {'name':'Deep Learning','price':1200,'category':'books','stock':10,'rating':4.5,'tags':['ai','machine learning']},

    'SKU012': {'name':'Chocolate','price':200,'category':'food','stock':100,'rating':4.3,'tags':['sweet','snack']},
    'SKU013': {'name':'Coffee','price':450,'category':'food','stock':60,'rating':4.4,'tags':['drink','energy']},
    'SKU014': {'name':'Pasta','price':300,'category':'food','stock':40,'rating':4.0,'tags':['italian','meal']},
    'SKU015': {'name':'Olive Oil','price':900,'category':'food','stock':20,'rating':4.2,'tags':['cooking','healthy']}
}


# Search by tag

def search_by_tag(tag):

    tag_groups = defaultdict(list)

    for sku, product in catalog.items():
        for t in product.get("tags", []):
            tag_groups[t].append(product)

    return tag_groups.get(tag, [])


# Out of stock products

def out_of_stock():

    return {sku:prod for sku,prod in catalog.items() if prod.get("stock",0)==0}


# Price range filter

def price_range(min_price,max_price):

    return {sku:prod for sku,prod in catalog.items()
            if min_price <= prod.get("price",0) <= max_price}


# Category summary

def category_summary():

    data = defaultdict(list)

    for product in catalog.values():

        category = product.get("category")
        data[category].append(product)

    summary = {}

    for category,products in data.items():

        count = len(products)

        avg_price = sum(p.get("price",0) for p in products)/count

        avg_rating = sum(p.get("rating",0) for p in products)/count

        summary[category] = {
            "count":count,
            "avg_price":avg_price,
            "avg_rating":avg_rating
        }

    return summary


# Apply discount

def apply_discount(category,percent):

    for product in catalog.values():

        if product.get("category")==category:

            product["price"] *= (1-percent/100)


# Merge catalogs

def merge_catalogs(catalog1,catalog2):

    return catalog1 | catalog2


# Example usage

print(search_by_tag("python"))
print(out_of_stock())
print(price_range(500,5000))
print(category_summary())
```

### 🚀 Part B — Log Analyzer using Counter & defaultdict

```python
# PART B — LOG ANALYZER

from collections import Counter, defaultdict


# Simulated logs

logs = [
"2026-03-10 INFO auth Login successful",
"2026-03-10 ERROR database Connection failed",
"2026-03-10 WARNING auth Password attempt failed",
"2026-03-10 ERROR database Query timeout",
"2026-03-10 INFO api Request received",
"2026-03-10 ERROR api Invalid request",
"2026-03-10 CRITICAL server System crash"
]


# Parse logs

parsed_logs = []

for log in logs:

    parts = log.split(" ",3)

    entry = {
        "timestamp":parts[0],
        "level":parts[1],
        "module":parts[2],
        "message":parts[3]
    }

    parsed_logs.append(entry)


# Counters

error_messages = Counter()
modules = Counter()
levels = Counter()

errors_by_module = defaultdict(list)

for entry in parsed_logs:

    level = entry["level"]
    module = entry["module"]
    message = entry["message"]

    levels[level]+=1
    modules[module]+=1

    if level=="ERROR":
        error_messages[message]+=1
        errors_by_module[module].append(message)


# Summary

summary = {

"total_entries":len(parsed_logs),

"error_rate":levels["ERROR"]/len(parsed_logs)*100,

"top_errors":error_messages.most_common(3),

"busiest_module":modules.most_common(1)[0][0]
}


print(summary)
print(errors_by_module)
```

### 🤖 Part D — AI Augmented Task (Improved Solution)

```python
# PART D — MERGE STUDENT GRADES ANALYSIS

from collections import defaultdict
from typing import Dict


def merge_grade_reports(sem1:Dict[str,float],sem2:Dict[str,float]):

    report = {}

    all_subjects = set(sem1) | set(sem2)

    for subject in all_subjects:

        g1 = sem1.get(subject)
        g2 = sem2.get(subject)

        if g1 and g2:

            combined = (g1+g2)/2

            if g2>g1:
                trend="improving"
            elif g2<g1:
                trend="declining"
            else:
                trend="stable"

        elif g1:
            combined=g1
            trend="single semester"

        else:
            combined=g2
            trend="single semester"

        report[subject] = {
            "semester1":g1,
            "semester2":g2,
            "combined_gpa":combined,
            "trend":trend
        }

    common_subjects = set(sem1) & set(sem2)

    return {
        "report":report,
        "common_subjects":common_subjects
    }


sem1 = {"math":3.5,"physics":3.8,"ai":3.9}
sem2 = {"math":3.7,"physics":3.6,"ml":4.0}

print(merge_grade_reports(sem1,sem2))
```
