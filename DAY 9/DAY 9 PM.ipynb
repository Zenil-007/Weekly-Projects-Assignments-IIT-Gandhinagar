### 🧠 Part A — Product Analytics Tool

```python
# PART A — PRODUCT ANALYTICS TOOL

from collections import namedtuple

# Define Named Tuple
Product = namedtuple("Product", ["id", "name", "category", "price"])


# Product Catalog (15 products)

p1 = Product(1, "Laptop", "Electronics", 75000)
p2 = Product(2, "Phone", "Electronics", 40000)
p3 = Product(3, "Tablet", "Electronics", 30000)
p4 = Product(4, "Headphones", "Electronics", 5000)

p5 = Product(5, "T-Shirt", "Clothing", 1000)
p6 = Product(6, "Jeans", "Clothing", 2500)
p7 = Product(7, "Jacket", "Clothing", 4500)
p8 = Product(8, "Sneakers", "Clothing", 3500)

p9 = Product(9, "Clean Code", "Books", 700)
p10 = Product(10, "Python Crash Course", "Books", 900)
p11 = Product(11, "Deep Learning", "Books", 1200)

p12 = Product(12, "Mixer", "Home", 3000)
p13 = Product(13, "Vacuum Cleaner", "Home", 8000)
p14 = Product(14, "Air Fryer", "Home", 6000)
p15 = Product(15, "Coffee Maker", "Home", 4500)

catalog = {p1,p2,p3,p4,p5,p6,p7,p8,p9,p10,p11,p12,p13,p14,p15}


# Customer carts (sets)

customer_1_cart = {p1, p5, p9, p12}
customer_2_cart = {p1, p6, p9, p13}
customer_3_cart = {p2, p5, p10, p12}
customer_4_cart = {p1, p7, p9, p14}
customer_5_cart = {p1, p8, p11, p12}

all_carts = [
    customer_1_cart,
    customer_2_cart,
    customer_3_cart,
    customer_4_cart,
    customer_5_cart
]


# Bestsellers (intersection)

bestsellers = set.intersection(*all_carts)
print("Bestsellers:", bestsellers)


# Catalog reach (union)

catalog_reach = set.union(*all_carts)
print("Products appearing in any cart:", catalog_reach)


# Exclusive purchases (customer 1)

others = set.union(customer_2_cart, customer_3_cart, customer_4_cart, customer_5_cart)
exclusive_customer1 = customer_1_cart - others

print("Products only customer 1 bought:", exclusive_customer1)


# Recommendation system

def recommend_products(customer_cart, all_carts):

    other_products = set.union(*all_carts)

    recommendations = other_products - customer_cart

    return recommendations


print("Recommendations for customer 1:")
print(recommend_products(customer_1_cart, all_carts))


# Category summary

def category_summary():

    categories = {p.category for p in catalog}

    summary = {
        cat: {p.name for p in catalog if p.category == cat}
        for cat in categories
    }

    return summary


print("Category Summary:")
print(category_summary())
```

### 🚀 Part B — Frozenset Bundle System

```python
# PART B — FROZENSET BUNDLE SYSTEM

"""
What is frozenset?

frozenset is an immutable version of a Python set.

Difference between set and frozenset:

set
- mutable
- supports add/remove operations
- cannot be dictionary keys

frozenset
- immutable
- cannot modify elements
- can be used as dictionary keys

Real-world usage:

Bundle pricing systems
Caching
Dictionary keys for complex objects
"""


# Bundle discount dictionary

bundle_discounts = {
    frozenset({'Electronics','Books'}): 10,
    frozenset({'Clothing','Home'}): 15,
    frozenset({'Electronics','Home'}): 12
}


# Bundle checker function

def check_bundle_discount(cart):

    cart_categories = {p.category for p in cart}

    for bundle, discount in bundle_discounts.items():

        if bundle.issubset(cart_categories):
            return discount

    return 0


print("Bundle Discount:", check_bundle_discount(customer_1_cart), "%")


# Performance benchmark

import timeit

set_time = timeit.timeit("set([1,2,3,4,5])", number=100000)
frozenset_time = timeit.timeit("frozenset([1,2,3,4,5])", number=100000)

print("Set creation time:", set_time)
print("Frozenset creation time:", frozenset_time)
```

### 🤖 Part D — AI Jaccard Similarity

```python
# PART D — AI JACCARD SIMILARITY

"""
Prompt Used:

Write a Python function that calculates the Jaccard similarity between two sets of strings.
Explain what Jaccard similarity is and where it is used in industry.
"""


# AI Generated Code

def jaccard_similarity(set_a, set_b):

    intersection = set_a.intersection(set_b)
    union = set_a.union(set_b)

    return len(intersection) / len(union)


# Test

set_a = {'python','java','sql'}
set_b = {'python','sql','docker','aws'}

print("Jaccard Similarity:", jaccard_similarity(set_a, set_b))


# Improved version (handles edge case)

def jaccard_similarity_safe(set_a, set_b):

    if not set_a and not set_b:
        return 1

    intersection = set_a & set_b
    union = set_a | set_b

    return len(intersection) / len(union)


print("Improved Similarity:", jaccard_similarity_safe(set_a,set_b))


"""
What is Jaccard Similarity?

Jaccard similarity measures how similar two sets are.

Formula:

|A ∩ B| / |A ∪ B|

Industry Uses:

Recommendation systems (similar user interests)
NLP document similarity
Plagiarism detection
Search engine ranking
"""
```
