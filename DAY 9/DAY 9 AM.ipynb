### 🧠 Part A — student_system.py

```python
# student_system.py

records = [
    ["Aman", "Math", 88],
    ["Priya", "Physics", 91],
    ["Rahul", "Math", 76],
    ["Sneha", "Chemistry", 85],
    ["Arjun", "Physics", 79],
    ["Kiran", "Math", 92],
    ["Neha", "Chemistry", 73],
    ["Vikas", "Physics", 84],
    ["Riya", "Math", 90],
    ["Ankit", "Chemistry", 81]
]


def add_student(name, subject, marks):

    for r in records:
        if r[0] == name and r[1] == subject:
            print("Record already exists!")
            return

    records.append([name, subject, marks])
    print("Student added successfully")


def get_toppers(subject):

    subject_students = [r for r in records if r[1] == subject]

    toppers = sorted(subject_students, key=lambda x: x[2], reverse=True)

    return toppers[:3]


def class_average(subject):

    marks = [r[2] for r in records if r[1] == subject]

    if len(marks) == 0:
        return 0

    return sum(marks) / len(marks)


def above_average_students():

    all_marks = [r[2] for r in records]

    avg = sum(all_marks) / len(all_marks)

    return [r for r in records if r[2] > avg]


def remove_student(name):

    global records

    records = [r for r in records if r[0] != name]

    print("Student removed if existed")


def save_records():

    with open("students.txt", "w") as f:

        for r in records:
            f.write(f"{r[0]}, {r[1]}, {r[2]}\n")


while True:

    print("\n--- Student Management System ---")

    print("1 Add student")
    print("2 Show toppers")
    print("3 Show class average")
    print("4 Show above-average students")
    print("5 Remove student")
    print("6 Exit")

    choice = input("Enter choice: ")

    if choice == "1":

        name = input("Enter name: ")
        subject = input("Enter subject: ")
        marks = int(input("Enter marks: "))

        add_student(name, subject, marks)

    elif choice == "2":

        subject = input("Enter subject: ")

        print(get_toppers(subject))

    elif choice == "3":

        subject = input("Enter subject: ")

        print("Average:", class_average(subject))

    elif choice == "4":

        print(above_average_students())

    elif choice == "5":

        name = input("Enter name to remove: ")

        remove_student(name)

    elif choice == "6":

        save_records()

        print("Records saved to students.txt")

        break

    else:

        print("Invalid choice")
```

### 🚀 Part B — matrix_ops.py

```python
# matrix_ops.py


def matrix_add(A, B):

    result = []

    for i in range(len(A)):
        row = []

        for j in range(len(A[0])):
            row.append(A[i][j] + B[i][j])

        result.append(row)

    return result


def matrix_transpose(matrix):

    return [[row[i] for row in matrix] for i in range(len(matrix[0]))]


def matrix_multiply(A, B):

    if len(A[0]) != len(B):
        return "Matrix multiplication not possible"

    result = [
        [
            sum(a * b for a, b in zip(row_a, col_b))
            for col_b in zip(*B)
        ]
        for row_a in A
    ]

    return result


if __name__ == "__main__":

    a = [[1, 2], [3, 4]]
    b = [[5, 6], [7, 8]]

    print(matrix_add(a, b))

    print(matrix_transpose(a))

    print(matrix_multiply(a, b))
```

### 💼 Part C — Interview Questions

### 1 Shallow Copy vs Deep Copy

Shallow Copy

A shallow copy creates a new object but references the same nested objects.

```python
import copy

a = [[1,2],[3,4]]

b = copy.copy(a)

b[0][0] = 99

print(a)
```

### Deep Copy

A deep copy creates a completely independent copy including nested objects.

```python
import copy

a = [[1,2],[3,4]]

b = copy.deepcopy(a)

b[0][0] = 99

print(a)
```
