# Clustering Assignment: K-Means, DBSCAN, Hierarchical Clustering

```

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score, silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
```

## Load and Preprocess Data

```

iris = load_iris()
X = iris.data
y = iris.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## K-Means Clustering

```

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

ari = adjusted_rand_score(y, kmeans_labels)
nmi = normalized_mutual_info_score(y, kmeans_labels)

print("ARI:", ari)
print("NMI:", nmi)
```

## Visualization

```

plt.figure()
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y)
plt.title("True Labels")

plt.figure()
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans_labels)
plt.title("KMeans Labels")
```

## DBSCAN

```

dbscan = DBSCAN(eps=0.5, min_samples=5)
db_labels = dbscan.fit_predict(X_scaled)

print("DBSCAN unique labels:", np.unique(db_labels))
```

## Hierarchical Clustering

```

agg = AgglomerativeClustering(n_clusters=3)
agg_labels = agg.fit_predict(X_scaled)

ari_agg = adjusted_rand_score(y, agg_labels)
print("Agglomerative ARI:", ari_agg)
```

```

linked = linkage(X_scaled, 'ward')

plt.figure()
dendrogram(linked)
plt.title("Dendrogram")
```

## KMeans From Scratch

```

def kmeans_from_scratch(X, k, max_iter=100):
    np.random.seed(42)
    centroids = X[np.random.choice(len(X), k, replace=False)]

    for _ in range(max_iter):
        distances = np.linalg.norm(X[:, None] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)

        new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])

        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    return labels, centroids

labels_scratch, centroids = kmeans_from_scratch(X_scaled, 3)
print("Scratch KMeans ARI:", adjusted_rand_score(y, labels_scratch))
```
